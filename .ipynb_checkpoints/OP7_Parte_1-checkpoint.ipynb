## Separación para entrenamiento y testeo

```python
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=2023)
```

```python
#@title Escalado de variables
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
#@title Modelo base
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(10, 64)  # Capa de entrada con 10 características
        self.fc2 = nn.Linear(64, 32)  # Capa oculta
        self.fc3 = nn.Linear(32, 1)   # Capa de salida

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
X_train = torch.Tensor(X_train)
X_test = torch.Tensor(X_test)
y_train = torch.Tensor(np.array(y_train))
y_test = torch.Tensor(np.array(y_test))
```

```python
model = Net()
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
error = nn.MSELoss()
```

```python
learning_rate = 0.02 ## Ver para parametrizar
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
num_epochs = 1000
```

```python
total = 0
correct = 0
accuracy_list_val = []
for epoch in range(num_epochs):
    outputs = model(X_train)
    loss = error(outputs, y_train.view(-1, 1))
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    predicted = torch.max(outputs.data, 1)[1]   # etiqueta predicha (WTA)
    total += len(y_train)                        # número total de etiquetas en lote
    correct += (predicted == y_train).sum()      # número de predicciones correctas
    if (epoch + 1) % 50 == 0:
      print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')
    # calcula el desempeño: Precisión (accuracy)
accuracy_val = float(correct) / float(total)
accuracy_list_val.append(accuracy_val)

print('Epoch: {:02}  Loss: {:.6f}  Accuracy: {:.6f}  Accuracy Val: {:.6f}'.format(epoch, loss.data, accuracy_val, accuracy_val))
```

```python
total = 0
correct= 0
accuracy_list_val = []
model.eval()
with torch.no_grad():

    y_pred = model(X_test)

    test_loss = error(y_pred, y_test.view(-1, 1))

    print(f'Loss en el conjunto de prueba: {test_loss.item():.4f}')
    predicted = torch.max(y_pred.data, 1)[1]   # etiqueta predicha (WTA)
    total += len(y_test)                        # número total de etiquetas en lote
    correct += (predicted == y_test).sum()      # número de predicciones correctas
accuracy_val = float(correct) / float(total)
accuracy_list_val.append(accuracy_val)

print('Epoch: {:02}  Loss: {:.6f}  Accuracy: {:.6f}'.format(epoch, loss.data, accuracy_val))
```

## Ajuste de Hiperparámetros y Mejora del Modelo

Para mejorar el rendimiento del modelo, se pueden realizar ajustes en los hiperparámetros y la arquitectura del modelo. A continuación, se presenta un ejemplo de cómo modificar el código existente para experimentar con diferentes configuraciones:


```python
import torch.optim as optim

# Ajuste de Hiperparámetros
learning_rate = 0.001
num_epochs = 1000

# Arquitectura del Modelo
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(10, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 1)
        self.dropout = nn.Dropout(0.2) 

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.dropout(x)
        x = torch.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.fc3(x)
        return x

model = Net()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.BCEWithLogitsLoss()  

for epoch in range(num_epochs):
    outputs = model(X_train)
    loss = criterion(outputs, y_train.view(-1, 1))
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 50 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

# Evaluación en conjunto de prueba
model.eval()
with torch.no_grad():
    y_pred = model(X_test)
    test_loss = criterion(y_pred, y_test.view(-1, 1))
    print(f'Loss en el conjunto de prueba: {test_loss.item():.4f}')
```
